```python
happy_eupho=happy_eupho.drop(["track"],axis=1)
```

```python
#fill missing null values
happy_eupho["danceability"].fillna(happy_eupho["danceability"].mean(),inplace=True)
happy_eupho["energy"].fillna(happy_eupho["energy"].median(),inplace=True)
happy_eupho["loudness"].fillna(happy_eupho["loudness"].mean(),inplace=True)
happy_eupho["mode"].fillna(happy_eupho["mode"].mean(),inplace=True)
happy_eupho["speechiness"].fillna(happy_eupho["speechiness"].median(),inplace=True)
happy_eupho["acousticness"].fillna(happy_eupho["acousticness"].median(),inplace=True)
happy_eupho["instrumentalness"].fillna(happy_eupho["instrumentalness"].mean(),inplace=True)
happy_eupho["liveness"].fillna(happy_eupho["liveness"].median(),inplace=True)
happy_eupho["valence"].fillna(happy_eupho["valence"].median(),inplace=True)
happy_eupho["tempo"].fillna(happy_eupho["tempo"].median(),inplace=True)
happy_eupho["duration_ms"].fillna(happy_eupho["duration_ms"].median(),inplace=True)
happy_eupho["time_signature"].fillna(happy_eupho["time_signature"].mean(),inplace=True)
happy_eupho["key"].fillna(happy_eupho["key"].mean(),inplace=True)
```

```python
happy_eupho.isnull().sum()
```

```python
happy_eupho['danceability']=happy_eupho['danceability'].astype(int)
```

```python
happy_eupho['mode']=happy_eupho['mode'].astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(happy_eupho.drop("mode", axis=1))
```

```python
scaled_features = scaler.transform(happy_eupho.drop('mode',axis=1))
scaled_features
```

```python
happy_eupho_feat = happy_eupho.drop("mode",axis=1)
happy_eupho_feat.head()
```

```python
from sklearn.model_selection import train_test_split
X = happy_eupho_feat
y = happy_eupho['mode'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=30, shuffle=True)
```

```python
y_test
```

#KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3,p=2,metric='euclidean')
knn.fit(X_train, y_train)
```

```python
prediction = knn.predict(X_test)
prediction
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction))
```

#SVM

```python
from sklearn import svm
clf= svm.SVC(kernel='poly')
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print("accuracy score",accuracy_score(y_test,y_pred))
```

```python
from sklearn import svm
clf= svm.SVC(kernel='rbf')
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print("accuracy score",accuracy_score(y_test,y_pred))
```

#Decision Tree

```python
from sklearn import tree
import graphviz
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=5)
dt.fit(X_train,y_train)
```

```python
y_test_pred=dt.predict(X_test)
accuracy_score(y_test,y_test_pred)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dt,
                   feature_names=X.columns,
                   class_names=["danceability","energy",	"key"	,"loudness","speechiness"	,"acousticness"	,"instrumentalness"	,"liveness"	,"valence"	,"tempo"	,"duration,_ms"	,"time_signature"],
                   filled=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)
y_pred_1 = clf.predict(X_test)
```
